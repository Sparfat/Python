## Multi-índice e hierarquia de índices

```python
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1, 2, 3, 1, 2, 3]
```

```python
hier_index = list(zip(outside, inside))
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6, 2), index = hier_index, columns = ['A', 'B'])
```

```python
df
```

```python
df.loc['G1']
```

# Groupby

```python
df = pd.DataFrame(data)
```

```python
df.groupby('Company').count()
```

# Mesclando, unindo e concatenando

```python
df1 = pd.DataFrame({'A' : ['A0', 'A1', 'A2', 'A3'], 'B' : ['B0', 'B1', 'B2', 'B3'], 'C' : ['C0', 'C1', 'C2', 'C3'], 'D' : ['D0', 'D1', 'D2', 'D3']}, index = [0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A' : ['A4', 'A5', 'A6', 'A7'], 'B' : ['B4', 'B5', 'B6', 'B7'], 'C' : ['C4', 'C5', 'C6', 'C7'], 'D' : ['D4', 'D5', 'D6', 'D7']}, index = [4, 5, 6, 7])
df2
```

```python
df3 = pd.DataFrame({'A' : ['A8', 'A9', 'A10', 'A11'], 'B' : ['B8', 'B9', 'B10', 'B11'], 'C' : ['C8', 'C9', 'C10', 'C11'], 'D' : ['D8', 'D9', 'D10', 'D11']}, index = [8, 9, 10, 11])
df3
```

## Concatenação

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis = 1)
```

## Mesclando

```python
esq = pd.DataFrame({'key' : ['K0', 'K1', 'K2', 'K3'], 'A' : ['A0', 'A1', 'A2', 'A3'], 'B' : ['B0', 'B1', 'B2', 'B3']})
esq
```

```python
dir = pd.DataFrame({'key' : ['K0', 'K1', 'K2', 'K3'], 'C' : ['C0', 'C1', 'C2', 'C3'], 'D' : ['D0', 'D1', 'D2', 'D3']})
dir
```

```python
pd.merge(esq, dir, how = 'inner', on = 'key')
```

```python
esq = pd.DataFrame({'key1' : ['K0', 'K0', 'K1', 'K2'], 'key2' : ['K0', 'K1', 'K0', 'K1'], 'A' : ['A0', 'A1', 'A2', 'A3'], 'B' : ['B0', 'B1', 'B2', 'B3']})
esq
```

```python
dir = pd.DataFrame({'key1' : ['K0', 'K1', 'K1', 'K2'], 'key2' : ['K0', 'K0', 'K0', 'K0'], 'C' : ['C0', 'C1', 'C2', 'C3'], 'D' : ['D0', 'D1', 'D2', 'D3']})
dir
```

```python
pd.merge(esq, dir, on = ['key1', 'key2'])
```

```python
pd.merge(esq, dir, how = 'outer', on = ['key1', 'key2'])
```

```python
pd.merge(esq, dir, how = 'right', on = ['key1', 'key2'])
```

```python
pd.merge(esq, dir, how = 'left', on = ['key1', 'key2'])
```

## Junção

```python
esq = pd.DataFrame({'A' : ['A0', 'A1', 'A2'], 'B' : ['B0', 'B1', 'B2']}, index = ['K0', 'K1', 'K2'])
esq
```

```python
dir = pd.DataFrame({'C' : ['C0', 'C2', 'C3'], 'D' : ['D0', 'D2', 'D3']}, index = ['K0', 'K2', 'K3'])
dir
```

```python
esq.join(dir)
```

```python
esq.join(dir, how = 'outer')
```
